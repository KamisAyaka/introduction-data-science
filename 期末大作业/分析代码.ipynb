```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('2018.csv')
df
```

```python
df = np.array(df)
team_name = df[:,0]
salary = df[:,1]
c = df[:,2]
d = df[:,3]
a = np.arange(30)

colors = []
for i in range(len(c)):
    if c[i] == 0:
        colors.append('g')
    else:
        colors.append('y')
        
for i in range(len(d)): 
    if d[i] == 1:
        colors[i] = 'r'

plt.figure(dpi=300,figsize=(24,8))
plt.bar(a,salary,label = 'salary',width=0.5, align='edge',color = colors)
plt.xlabel('team name',fontsize='15')
plt.ylabel('salary',fontsize='15')
plt.xticks(a,team_name,rotation = 60,fontsize = '15')
plt.title('2018-2019 nba playoffs team salary',fontsize = '20')
plt.legend()
plt.show()
```

```python
df1 = pd.read_csv('2019.csv')
df1 = np.array(df1)
team_name1 = df1[:,0]
c1 = df1[:,2]

for i in range(len(c1)):
    name = team_name1[i]
    for j in range(len(c1)):
        if name == team_name[j]:
            c[j] = c[j] + c1[i]
            break
     
    
df1 = pd.read_csv('2020.csv')
df1 = np.array(df1)
team_name1 = df1[:,0]
c1 = df1[:,2]

for i in range(len(c1)):
    name = team_name1[i]
    for j in range(len(c1)):
        if name == team_name[j]:
            c[j] = c[j] + c1[i]
            break
        
            
df1 = pd.read_csv('2021.csv')
df1 = np.array(df1)
team_name1 = df1[:,0]
c1 = df1[:,2]

for i in range(len(c1)):
    name = team_name1[i]
    for j in range(len(c1)):
        if name == team_name[j]:
            c[j] = c[j] + c1[i]
            break         
            
            
df1 = pd.read_csv('2022.csv')
df1 = np.array(df1)
team_name1 = df1[:,0]
c1 = df1[:,2]

for i in range(len(c1)):
    name = team_name1[i]
    for j in range(len(c1)):
        if name == team_name[j]:
            c[j] = c[j] + c1[i]
            break
```

```python
plt.figure(dpi=300,figsize=(24,8))
plt.bar(a,c,label = 'count',width=0.5, align='edge')
plt.xticks(a,team_name,rotation = 60,fontsize = '15')
plt.legend()
plt.show()
```
